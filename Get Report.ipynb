```python
# Baca Token
import json
# set sektor dan tahun
sektor = 'IDXENERGY'
tahun = '2017'
path = '/content/drive/MyDrive/Skripsi_Putri/Hasil/Tokens/%s_%s.json'%(sektor,tahun)
with open(path, 'r') as fp:
  pdf_tokens = json.load(fp)
company = list(pdf_tokens.keys())
```

```python
#Baca hasil similarity
dataset = dict()
for comp in company :
  try :
    with open('/content/drive/MyDrive/Skripsi_Putri/Hasil/Similarity/%s/%s_%s.json'%(sektor,comp,tahun), 'r') as fp:
      dataset.update({comp : json.load(fp)})
  except :
    print('error',sektor,comp)
    break
```

```python
#method untuk eliminasi dibawah treshold/param
def elimination(sim_res, param=0.75) :
  res = sim_res.copy()
  dict_result = dict()
  for comp in res.keys() :
    dict_temp = dict()
    for word in res[comp].keys() :
      if (len(word.split(' ')) != len(res[comp][word][1].split())) :
        continue
      if (res[comp][word][2] >= param) :
        dict_temp[word] = res[comp][word]
    dict_result[comp] = dict_temp
  return dict_result
  

#method untuk mendapatkan report unique
def getUniqueReport(data) :
  results = data.copy()
  report = dict()
  listWord = dict()
  for comp in results.keys() :
    innovation, integrity, quality, respect, teamwork = 0,0,0,0,0
    listInnovation, listIntegrity, listQuality, listRespect, listTeamwork = list(),list(),list(),list(),list()
    for key in results[comp].keys() :
      if results[comp][key][0] == 'Innovation' :
        innovation += 1
        listInnovation.append(key)
      elif results[comp][key][0] == 'Integrity' :
        integrity += 1
        listIntegrity.append(key)
      elif results[comp][key][0] == 'Quality' :
        quality += 1
        listQuality.append(key)
      elif results[comp][key][0] == 'Respect' :
        respect += 1
        listRespect.append(key)
      elif results[comp][key][0] == 'Teamwork' :
        teamwork += 1
        listTeamwork.append(key)
    report[comp] = {'Innovation' : innovation,
                    'Integrity' : integrity,
                    'Quality' : quality,
                    'Respect' : respect,
                    'Teamwork' : teamwork}
    listWord[comp] = {'Innovation' : sorted(set(listInnovation)),
                      'Integrity' : sorted(set(listIntegrity)),
                      'Quality' : sorted(set(listQuality)),
                      'Respect' : sorted(set(listRespect)),
                      'Teamwork' : sorted(set(listTeamwork))}
  return report,listWord

#method untuk mendapatkan report
def ReportAll(listWord) :
  res = dict()
  report = dict()
  for comp in listWord :
    res[comp] = dict()
    report[comp] = dict()
    for cult in listWord[comp] :
      res[comp][cult] = dict()
      report[comp][cult] = 0
      for word in listWord[comp][cult] :
        word_count = pdf_tokens[comp].count(word)
        res[comp][cult][word] = word_count
        report[comp][cult] += word_count
  return report,res
```

```python
token_selection = elimination(dataset)
ReportUnique,listWordUnique = getUniqueReport(token_selection)
Report, count_word = ReportAll(listWordUnique)
```
