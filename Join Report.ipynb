```python
import pandas as pd
import numpy as np
import json
```

```python
data = pd.read_csv('/content/drive/MyDrive/Skripsi_Putri/DataEmiten_2017.csv')
data.drop([0],axis=0,inplace=True)
data = data[data.columns[:4]]
# data.head()
sektor = data['Sektor'].unique()
sektor = list(sektor)
sektor.remove('DELISTING')
```

```python
data.head()
```

```python
tahun = '2017'
join_report = []
join_list = []
join_count = []
join_unique = []

for sek in sektor :
  Report = pd.read_csv("/content/drive/MyDrive/Skripsi_Putri/Hasil/Report/%s/Report_%s_%s.csv"%(sek,sek,tahun))
  Report.rename(columns = {"Unnamed: 0" : "Code"},inplace=True)
  join_report.append(Report)

  ListWord = pd.read_csv("/content/drive/MyDrive/Skripsi_Putri/Hasil/Report/%s/ListWord_%s_%s.csv"%(sek,sek,tahun))
  ListWord.rename(columns = {"Unnamed: 0" : "Code"},inplace=True)
  join_list.append(ListWord)

  CountWord = pd.read_csv("/content/drive/MyDrive/Skripsi_Putri/Hasil/Report/%s/CountWord_%s_%s.csv"%(sek,sek,tahun))
  CountWord.rename(columns = {"Unnamed: 0" : "Code"},inplace=True)
  join_count.append(CountWord)

  ReportUnique = pd.read_csv("/content/drive/MyDrive/Skripsi_Putri/Hasil/Report/%s/ReportUnique_%s_%s.csv"%(sek,sek,tahun))
  ReportUnique.rename(columns = {"Unnamed: 0" : "Code"},inplace=True)
  join_unique.append(ReportUnique)

Report = pd.concat(join_report)
Report.reset_index(inplace=True)
Report.drop('index',inplace=True,axis=1)

ListWord = pd.concat(join_list)
ListWord.reset_index(inplace=True)
ListWord.drop('index',inplace=True,axis=1)

CountWord = pd.concat(join_count)
CountWord.reset_index(inplace=True)
CountWord.drop('index',inplace=True,axis=1)

ReportUnique = pd.concat(join_unique)
ReportUnique.reset_index(inplace=True)
ReportUnique.drop('index',inplace=True,axis=1)
```

```python
Report = data.merge(Report, how='inner', on='Code')
Report.reset_index(inplace=True)
Report.drop('index',inplace=True,axis=1)
Report.tail()
```

```python
ListWord = data.merge(ListWord, how='inner', on='Code')
ListWord.reset_index(inplace=True)
ListWord.drop('index',inplace=True,axis=1)
ListWord.tail()
```

```python
CountWord = data.merge(CountWord, how='inner', on='Code')
CountWord.reset_index(inplace=True)
CountWord.drop('index',inplace=True,axis=1)
CountWord.tail()
```

```python
ReportUnique = data.merge(ReportUnique, how='inner', on='Code')
ReportUnique.reset_index(inplace=True)
ReportUnique.drop('index',inplace=True,axis=1)
ReportUnique.tail()
```

```python
#SAVE
Report.to_csv('/content/drive/MyDrive/Skripsi_Putri/Hasil/Join_Report_%s.csv'%tahun,index=False)
ListWord.to_csv('/content/drive/MyDrive/Skripsi_Putri/Hasil/Join_ListWord_%s.csv'%tahun,index=False)
CountWord.to_csv('/content/drive/MyDrive/Skripsi_Putri/Hasil/Join_CountWord_%s.csv'%tahun,index=False)
ReportUnique.to_csv('/content/drive/MyDrive/Skripsi_Putri/Hasil/Join_ReportUnique_%s.csv'%tahun,index=False)
```
