```python
#skip warning
import warnings
warnings.filterwarnings('ignore')
#connect to drive
from google.colab import drive  
import pandas as pd
import numpy as np
import json
drive.mount('/content/drive')
```

```python
tahun = '2017'
data = pd.read_csv('/content/drive/MyDrive/Skripsi_Putri/DataEmiten_%s.csv'%tahun)
data.drop([0],axis=0,inplace=True)
data = data[data.columns[:4]]
# data.head()
sektor = data['Sektor'].unique()
sektor = list(sektor)
sektor.remove('DELISTING')
sektor
Code = list(data['Code'])
```

```python
similarity = dict()
for sek in sektor :
  for code in Code :
    try :
      path = '/content/drive/MyDrive/Skripsi_Putri/Hasil/Similarity/%s/%s_%s.json'%(sek,code,tahun)
      with open(path, 'r') as fp:
        js = json.load(fp)
        similarity.update(js)
    except :
      continue
```

```python
list_word = pd.read_csv('/content/drive/MyDrive/Skripsi_Putri/Hasil/Unique_Word_Per_Cultures_%s.csv'%tahun)
```

```python
list_word.head()
```

```python
list_word['Innovation_Similarity'] = 0
list_word['Integrity_Similarity'] = 0
list_word['Quality_Similarity'] = 0
list_word['Respect_Similarity'] = 0
list_word['Teamwork_Similarity'] = 0
```

```python
for i in range(len(list_word)) :
  for col in list_word.columns :
    if type(list_word.loc[i,col]) == str :
      c_name = '%s_Similarity'%col
      list_word.loc[i,c_name] = similarity[list_word.loc[i,col]][2]
```

```python
list_word.head()
```

```python
for col in list(list_word.columns)[:5] :
  c_name = '%s_Similarity'%col
  list_word.loc[:,[col,c_name]] = list_word.loc[:,[col,c_name]].sort_values(c_name,ascending=False).values
```
