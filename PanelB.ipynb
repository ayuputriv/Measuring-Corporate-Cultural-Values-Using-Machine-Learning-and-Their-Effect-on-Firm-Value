```python
import pandas as pd
import json
#skip warning
import warnings
warnings.filterwarnings('ignore')
#connect to drive
from google.colab import drive  
import pandas as pd
import numpy as np
import json
drive.mount('/content/drive')
```

```python
def CustomParser(data):
    j1 = json.loads(data.replace("'",'"'))
    return j1
```

```python
tahun = '2017'
tf_idf = pd.read_csv('/content/drive/MyDrive/Skripsi_Putri/Hasil/TF_IDF_%s.csv'%tahun,
                        converters={'Innovation':CustomParser,
                                    'Integrity':CustomParser,
                                    'Quality':CustomParser,
                                    'Respect':CustomParser,
                                    'Teamwork':CustomParser})
tf_idf.head()
```

```python
word = pd.read_csv('/content/drive/MyDrive/Skripsi_Putri/Hasil/Unique_Word_Per_Cultures_%s.csv'%tahun)
word.head()
```

```python
cultures = list(word.columns)
word['Innovation_tf_idf_sum'] = 0
word['Integrity_tf_idf_sum'] = 0
word['Quality_tf_idf_sum'] = 0
word['Respect_tf_idf_sum'] = 0
word['Teamwork_tf_idf_sum'] = 0
```

```python
Innovation = dict(pd.DataFrame(list(tf_idf.loc[:,'Innovation'])).sum())
Integrity = dict(pd.DataFrame(list(tf_idf.loc[:,'Integrity'])).sum())
Quality = dict(pd.DataFrame(list(tf_idf.loc[:,'Quality'])).sum())
Respect = dict(pd.DataFrame(list(tf_idf.loc[:,'Respect'])).sum())
Teamwork = dict(pd.DataFrame(list(tf_idf.loc[:,'Teamwork'])).sum())
```

```python
for i in range(len(word)) :
  if type(word.loc[i,'Innovation']) == str :
    word.loc[i,'Innovation_tf_idf_sum'] = Innovation[word.loc[i,'Innovation']]
  if type(word.loc[i,'Integrity']) == str :
    word.loc[i,'Integrity_tf_idf_sum'] = Integrity[word.loc[i,'Integrity']]
  if type(word.loc[i,'Quality']) == str :
    word.loc[i,'Quality_tf_idf_sum'] = Quality[word.loc[i,'Quality']]
  if type(word.loc[i,'Respect']) == str :
    word.loc[i,'Respect_tf_idf_sum'] = Respect	[word.loc[i,'Respect']]
  if type(word.loc[i,'Teamwork']) == str :
    word.loc[i,'Teamwork_tf_idf_sum'] = Teamwork[word.loc[i,'Teamwork']]
```

```python
word.head()
```

```python
for col in list(word.columns)[:5] :
  c_name = '%s_tf_idf_sum'%col
  word.loc[:,[col,c_name]] = word.loc[:,[col,c_name]].sort_values(c_name,ascending=False).values
```

```python
word.head()
```

```python
cols = list(word.columns)
cols.sort()
word = word[cols]
word.head()
```

```python
word.to_csv("/content/drive/MyDrive/Skripsi_Putri/Hasil/Panel_B_sum_%s"%tahun,index=False)
```
