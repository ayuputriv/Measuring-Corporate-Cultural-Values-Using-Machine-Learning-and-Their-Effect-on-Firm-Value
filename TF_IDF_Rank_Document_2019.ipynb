```python
def CustomParser(data):
    j1 = json.loads(data.replace("'",'"'))
    return j1
```

```python
WordCount = pd.read_csv("/content/drive/MyDrive/Skripsi_Putri/Hasil Akhir/Hasil_Gabungan_2.csv",
                        converters={'Innovation_word':CustomParser,
                                    'Integrity_word':CustomParser,
                                    'Quality_word':CustomParser,
                                    'Respect_word':CustomParser,
                                    'Teamwork_word':CustomParser})
WordCount.head()
```

```python
word = pd.read_csv('/content/drive/MyDrive/Skripsi_Putri/Hasil Akhir/Word List Unique.csv')
word.rename(columns = {"Repsect" : "Respect"},inplace=True)
word.head()
```

```python
WordCount[WordCount.Sektor == 'DELISTING']
```

```python
WordCount = WordCount[WordCount.Sektor != 'DELISTING']
WordCount.reset_index(drop=True,inplace=True)
WordCount.tail()
```

```python
word2 = []
word2 +=list(word['Innovation'].dropna())
word2 += list(word['Integrity'].dropna())
word2 += list(word['Quality'].dropna())
word2 += list(word['Respect'].dropna())
word2 += list(word['Teamwork'].dropna())
len(word2)
```

```python
df_res = pd.DataFrame(word2,columns=['word'])
df_res['df'] = 0
df_res.head()
```

```python
for i in range(len(df_res)) :
  for j in range(len(WordCount)) :
    cek = False
    for k in WordCount.columns[7:] :
      if df_res.iloc[i,0] in WordCount.loc[j,k].keys() :
        cek = True
        break
    if cek :
      df_res.iloc[i,1] += 1
```

```python
df_res.head()
```

```python
import math
def idf(x) :
  return math.log(len(WordCount)/x)
df_res['idf'] = df_res['df'].apply(idf)
df_res.head()
```

```python
df_res.sort_values('idf',inplace=True,ascending=False)
df_res.head()
```

```python
df_res.to_csv("/content/drive/MyDrive/Skripsi_Putri/Hasil Akhir/IDF_%s.csv"%2019,index=False)
```

```python
WordCount['Count All Word'] = 0
```

```python
for i in range(len(WordCount)) :
  sum_word = 0
  for j in WordCount.columns[2:7] :
    sum_word += WordCount.loc[i,j]
  WordCount.iloc[i,-1] = sum_word
```

```python
WordCount3 = WordCount2.copy()
WordCount3['Innovation_tf_idf_sum'] = 0
WordCount3['Integrity_tf_idf_sum'] = 0
WordCount3['Quality_tf_idf_sum'] = 0
WordCount3['Respect_tf_idf_sum'] = 0
WordCount3['Teamwork_tf_idf_sum'] = 0
for i in range(len(WordCount3)) :
  for j in WordCount3.columns[7:12] :
    WordCount3.loc[i,j[:-5]+'_tf_idf_sum'] = sum(WordCount3.loc[i,j].values())
```
