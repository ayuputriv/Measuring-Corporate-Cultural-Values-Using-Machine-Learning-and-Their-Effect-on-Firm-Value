```python
#skip warning
import warnings
warnings.filterwarnings('ignore')
#connect to drive
from google.colab import drive  
import pandas as pd
import numpy as np
import json
drive.mount('/content/drive')
```

```python
data = pd.read_csv('/content/drive/MyDrive/Skripsi_Putri/DataEmiten_2017.csv')
data.drop([0],axis=0,inplace=True)
data = data[data.columns[:4]]
# data.head()
sektor = data['Sektor'].unique()
sektor = list(sektor)
sektor.remove('DELISTING')
sektor
```

```python
kata = pd.read_csv('/content/drive/MyDrive/Skripsi_Putri/Data_kata.csv')
for col in kata.columns :
  kata[col] = kata[col].str.replace('_',' ')
  kata[col] = kata[col].str.replace('-','')
  kata[col] = kata[col].str.lower()
kata.head()
```

```python
tahun = '2017'
res_sektor = []
for sek in sektor :
  Report = pd.read_csv("/content/drive/MyDrive/Skripsi_Putri/Hasil/Report/%s/Report_%s_%s.csv"%(sek,sek,tahun))
  Report.rename(columns = {"Unnamed: 0" : "Code"},inplace=True)
  res_sektor.append(Report)
Report = pd.concat(res_sektor)
Report.reset_index(inplace=True)
Report.drop('index',inplace=True,axis=1)
Report
```

```python
def CustomParser(data):
    j1 = json.loads(data.replace("'",'"'))
    return j1
```

```python
count_data = []
for sek in sektor :
  CountWord = pd.read_csv("/content/drive/MyDrive/Skripsi_Putri/Hasil/Report/%s/CountWord_%s_%s.csv"%(sek,sek,tahun),
                        converters={'Innovation':CustomParser,
                                    'Integrity':CustomParser,
                                    'Quality':CustomParser,
                                    'Respect':CustomParser,
                                    'Teamwork':CustomParser})
  CountWord.rename(columns = {"Unnamed: 0" : "Code"},inplace=True)
  count_data.append(CountWord)
CountWord = pd.concat(count_data)
CountWord.reset_index(inplace=True)
CountWord.drop('index',inplace=True,axis=1)
CountWord
```

```python
CountWord2 = CountWord.set_index('Code')
CountWord2.head()
```

```python
ListWord = {'Innovation' : set(),
            'Integrity' : set(),
            'Quality' : set(),
            'Respect' : set(),
            'Teamwork' : set()}
for i in range(len(CountWord)) :
  for j in CountWord.columns[1:] :
    ListWord[j].update(set(CountWord.loc[i,j].keys()))
```

```python
for k in ListWord :
  ListWord[k] = list(ListWord[k])
  ListWord[k].sort()
UniqueListWordAll = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in ListWord.items()]))
```

```python
Count_Per_Sektor = {'Innovation' : pd.DataFrame(columns=ListWord['Innovation']),
                    'Integrity' : pd.DataFrame(columns=ListWord['Integrity']),
                    'Quality' : pd.DataFrame(columns=ListWord['Quality']),
                    'Respect' : pd.DataFrame(columns=ListWord['Respect']),
                    'Teamwork' : pd.DataFrame(columns=ListWord['Teamwork'])}
```

```python
for key in Count_Per_Sektor :
  Count_Per_Sektor[key] = pd.concat([CountWord[CountWord.columns[:1]],Count_Per_Sektor[key]],axis=1)
  Count_Per_Sektor[key].fillna(0,inplace=True)
  Count_Per_Sektor[key].set_index('Code',inplace=True)
```

```python
for row in CountWord2.index :
  for col in list(CountWord2.columns) :
    for word in CountWord2.loc[row,col] :
      Count_Per_Sektor[col].loc[row,word] += CountWord2.loc[row,col][word] 
```

```python
similarity = dict()
for sek in sektor :
  for code in CountWord2.index :
    try :
      path = '/content/drive/MyDrive/Skripsi_Putri/Hasil/Similarity/%s/%s_%s.json'%(sek,code,tahun)
      with open(path, 'r') as fp:
        similarity.update(json.load(fp))
    except :
      continue
```

```python
kata_per_company = {'Innovation' : pd.DataFrame(columns=kata['Innovation'].dropna()),
                    'Integrity' : pd.DataFrame(columns=kata['Integrity'].dropna()),
                    'Quality' : pd.DataFrame(columns=kata['Quality'].dropna()),
                    'Respect' : pd.DataFrame(columns=kata['Respect'].dropna()),
                    'Teamwork' : pd.DataFrame(columns=kata['Teamwork'].dropna())}
```

```python
for key in kata_per_company :
  kata_per_company[key] = pd.concat([CountWord[CountWord.columns[:1]],kata_per_company[key]],axis=1)
  kata_per_company[key].fillna(0,inplace=True)
  kata_per_company[key].set_index('Code',inplace=True)
```

```python
for row in CountWord2.index :
  for col in list(CountWord2.columns) :
    for word in CountWord2.loc[row,col] :
      w = similarity[word][1]
      kata_per_company[col].loc[row,w] += CountWord2.loc[row,col][word] 
```

```python
#MERGE
for key in kata_per_company :
  Count_Per_Sektor[key] = data.merge(Count_Per_Sektor[key], how='inner', on='Code')
  Count_Per_Sektor[key].reset_index(inplace=True)
  Count_Per_Sektor[key].drop('index',inplace=True,axis=1)
```

```python
#SAVE
UniqueListWordAll.to_csv('/content/drive/MyDrive/Skripsi_Putri/Hasil/Unique_Word_Per_Cultures_%s.csv'%tahun,index=False)
for key in kata_per_company :
  Count_Per_Sektor[key].to_csv('/content/drive/MyDrive/Skripsi_Putri/Hasil/CountWord_Per_Company_%s_%s.csv'%(key,tahun))
  kata_per_company[key].to_csv('/content/drive/MyDrive/Skripsi_Putri/Hasil/CountSeedWord_Per_Company_%s_%s.csv'%(key,tahun))
```
